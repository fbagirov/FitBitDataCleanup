```python
import pandas as pd
pd.options.display.float_format = '{:,.2f}'.format
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
minuteSteps = pd.read_csv('minuteSteps.csv')
minuteSteps = minuteSteps.drop(u'Unnamed: 0', axis=1)
minuteSteps.head()
```

```python
minuteIntensities = pd.read_csv('minuteIntensities.csv')
minuteIntensities = minuteIntensities.drop(u'Unnamed: 0', axis=1)
minuteIntensities.head()
```

```python
dailyDistances = pd.read_csv('dailyDistances.csv')
dailyDistances = dailyDistances.drop(u'Unnamed: 0', axis=1)
dailyDistances.head()
```

```python
#merge minuteIntensities with minuteSteps
#the objective to get the steps where the person
#was active, and not steping in place
minuteIntensities_Steps = pd.merge(minuteIntensities,
                                   minuteSteps,
                                   on = ['Id', 
                                    'ActivityMinute'])
#extract only records when user was active
minuteIntensities_Steps_Active = minuteIntensities_Steps[
    minuteIntensities_Steps['Intensity'] > 0]
#add active steps per Id
TotalSteps_Active = minuteIntensities_Steps_Active.groupby(
    'Id')['Steps'].sum().reset_index()
#use dailyIntensities information to calculate active 
#distance per customer daily
dailyDistances['Distance'] = dailyDistances[
    ['LightActiveDistance', 'ModeratelyActiveDistance', 
     'VeryActiveDistance']].sum(axis=1)
#add active distances per Id
TotalDistance_Active = dailyDistances.groupby('Id')[
    'Distance'].sum().reset_index()
#merge active distances and steps per Id
DistanceSteps = pd.merge(TotalDistance_Active,
                         TotalSteps_Active,
                         on = ['Id'])
#calculate Height per Id
#The formula takes the number of feet in the mile
#(5280) and multiplies it by distance walked and 
#divides by number of steps multiplied by 
#the stride lengths to heigh coefficient  
DistanceSteps['Height'] = DistanceSteps.apply(
    lambda x: 5280*x.Distance/(x.Steps*0.65), axis=1)
#format Height to standard notation
DistanceSteps['Height Formatted'] = DistanceSteps[
    'Height'].apply(lambda x: str(int(np.floor(x)
    )) + '\' ' +  str(int(round(12*(x % 1)))) + '\"')
DistanceSteps
```

```python
def convert_id_to_int(id):
    try:
        return int(id)
    except ValueError:
        return int(id[1:])

minuteSteps_clean = minuteSteps.copy()
minuteSteps_clean['Id'] = minuteSteps_clean['Id'].apply(convert_id_to_int)
minuteSteps_clean.info()
```

One would not expect # of steps to be zero for intensities 2 and 3.
Impute Steps values as follows:
1-fill nulls with averages for that customer (excluding 500's and 0's)
2-replace 500 with average for that customer (excluding 500's and 0's)
3-replace 0's with intentisities 2 and 3 with averages for that customer (excluding 500's and 0's)

### Inpute values using different methods:

1. mean value
2. interpolate between existing values
3. draw from the distribution of existing values (per customer)

```python
#impute using mean
minuteIntensities_Steps_mean = minuteIntensities_Steps_clean.copy()

def compute_mean_imputed_values(data):
    return data[(data != 0) & (data != 500)].mean()

minuteSteps_means = minuteIntensities_Steps_mean.groupby('Id')[
          'Steps'].apply(compute_mean_imputed_values).to_dict()

minuteIntensities_Steps_mean.loc[
             (minuteIntensities_Steps_mean['Steps'] == 500) 
           | (pd.isnull(minuteIntensities_Steps_mean['Steps'])),
              'Steps'] = minuteIntensities_Steps_mean.loc[
             (minuteIntensities_Steps_mean['Steps'] == 500)
           | (pd.isnull(minuteIntensities_Steps_mean['Steps'])),
              'Id'].apply(lambda x: minuteSteps_means[x])
```

```python
#impute using interpolation
#assume the array is already sorted by time
minuteIntensities_Steps_interpolate = minuteIntensities_Steps_clean.copy()
minuteIntensities_Steps_interpolate['Steps'] = minuteIntensities_Steps_interpolate['Steps'].interpolate()
minuteIntensities_Steps_interpolate.info()
```

```python
#Impute using transform with random choice
minuteIntensities_Steps_distr = minuteIntensities_Steps_clean.copy()

def impute_from_distribution(data):
    values = data[data != 500].dropna().values
    #print values
    data.loc[(data == 500) | 
             (pd.isnull(data))
            ] = data.loc[(data == 500) | 
                         (pd.isnull(data))].apply(
        lambda x: np.random.choice(values))
    return data

minuteIntensities_Steps_distr.loc[
    minuteIntensities_Steps_clean['Intensity'] > 0, 'Steps'
    ] = minuteIntensities_Steps_distr[
    minuteIntensities_Steps_distr['Intensity'] > 0
    ].groupby('Id')['Steps'].transform(impute_from_distribution)
```

### Calculate Values using different imputed versions

```python
def calculate_height(minuteIntensities_Steps, dailyDistances):
    #extract only records when user was active
    minuteIntensities_Steps_Active = minuteIntensities_Steps[minuteIntensities_Steps['Intensity'] > 0]
    #add active steps per Id
    TotalSteps_Active = minuteIntensities_Steps_Active.groupby('Id')['Steps'].sum().reset_index()
    #use dailyIntensities information to calculate active distance per customer daily
    dailyDistances['Distance'] = dailyDistances[
        ['LightActiveDistance', 'ModeratelyActiveDistance', 'VeryActiveDistance']].sum(axis=1)
    #add active distances per Id
    TotalDistance_Active = dailyDistances.groupby('Id')['Distance'].sum().reset_index()
    #merge active distances and steps per customer Id
    DistanceSteps = pd.merge(TotalDistance_Active,
                             TotalSteps_Active,
                             on = ['Id'])
    #calculate Height per Id
    DistanceSteps['Height'] = DistanceSteps.apply(lambda x: 5280*x.Distance/(x.Steps*0.65), axis=1)
    #format Height to standard notation
    DistanceSteps['Height Formatted'] = DistanceSteps['Height'].apply(lambda x: 
                                                          str(int(np.floor(x))) + '\' ' +  str(int(round(12*(x % 1)))) + '\"')
    return DistanceSteps
```

```python
HeightMean = calculate_height(minuteIntensities_Steps_mean, dailyDistances)[['Id', 'Height Formatted']]
HeightMean.columns = ['Id', 'Height Mean']
HeightInterpolate = calculate_height(minuteIntensities_Steps_interpolate, dailyDistances)[['Id', 'Height Formatted']]
HeightInterpolate.columns = ['Id', 'Height Interpolate']
HeightDistribution = calculate_height(minuteIntensities_Steps_distr, dailyDistances)[['Id', 'Height Formatted']] 
HeightDistribution.columns = ['Id', 'Height Distribution']
```

### Calculate Original Height

```python
dailyIntensities_orig = pd.read_csv('data/dailyIntensities_merged.csv')
minuteIntensities_orig = pd.read_csv('data/minuteIntensitiesNarrow_merged.csv')
minuteSteps_orig = pd.read_csv('data/minuteStepsNarrow_merged.csv')
minuteIntensities_Steps_orig = pd.merge(minuteIntensities_orig,
                                   minuteSteps_orig,
                                   on = ['Id', 'ActivityMinute'])
minuteIntensities_Steps_Active_orig = minuteIntensities_Steps_orig[minuteIntensities_Steps_orig['Intensity'] > 0]
TotalSteps_Active_orig = minuteIntensities_Steps_Active_orig.groupby('Id')['Steps'].sum().reset_index()
dailyIntensities_orig['Distance'] = dailyIntensities_orig[
    ['LightActiveDistance', 'ModeratelyActiveDistance', 'VeryActiveDistance']].sum(axis=1)
TotalDistance_Active_orig = dailyIntensities_orig.groupby('Id')['Distance'].sum().reset_index()
DistanceSteps_orig = pd.merge(TotalDistance_Active_orig,
                         TotalSteps_Active_orig,
                         on = ['Id'])
DistanceSteps_orig['Height'] = DistanceSteps_orig.apply(lambda x: 5280*x.Distance/(x.Steps*0.65), axis=1)
DistanceSteps_orig['Height Formatted'] = DistanceSteps_orig['Height'].apply(lambda x: 
                                                      str(int(np.floor(x))) + '\' ' +  str(int(round(12*(x % 1)))) + '\"')
HeightOriginal = DistanceSteps_orig[['Id', 'Height Formatted']]
HeightOriginal.columns = ['Id', 'Height Original']
```

```python
HeightMerge = pd.merge(HeightOriginal,
                       HeightMean,
                       on = 'Id')

HeightMerge = pd.merge(HeightMerge,
                       DistanceSteps[['Id', 'Height Dirty']],
                       on = 'Id')

HeightMerge = pd.merge(HeightMerge,
                       HeightInterpolate,
                       on = 'Id')

HeightMerge = pd.merge(HeightMerge,
                    HeightDistribution,
                       on = 'Id')
                       
HeightMerge.loc[HeightMerge['Height Original'] != HeightMerge['Height Dirty'], 
                ['Id', 'Height Original', 'Height Dirty', 'Height Mean', 'Height Interpolate', 'Height Distribution']]         
```
